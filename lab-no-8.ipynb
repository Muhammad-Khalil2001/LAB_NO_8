```python
import cv2
import matplotlib.pyplot as plt
# Load an image from a file
image = cv2.imread('/kaggle/input/labno8/image.jpg')
# Convert the image from BGR to RGB (OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Display the image
plt.imshow(image_rgb)
plt.title('Loaded Image')
plt.axis('off') # Hide axis
plt.show()
```

```python
# Resize the image to 200x200 pixels
resized_image = cv2.resize(image_rgb, (200, 200))
# Display the resized image
plt.imshow(resized_image)
plt.title('Resized Image (200x200)')
plt.axis('off')
plt.show()
```

```python
# Apply horizontal flip
flipped_image = cv2.flip(image_rgb, 1) 
# Display the flipped image
plt.imshow(flipped_image) 
plt.title('Horizontally Flipped Image') 
plt.axis('off') 
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
# Use Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)
# Display the edges
plt.imshow(edges, cmap='gray')
plt.title('Edges in Image')
plt.axis('off')
plt.show()
```

```python
# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Display the histogram for the grayscale image
plt.hist(gray_image.ravel(), bins=256, range=[0,256])
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Display the histogram for each color channel (RGB)
colors = ('r', 'g', 'b')
for i, color in enumerate(colors):
 histogram = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
 plt.plot(histogram, color=color)
 plt.xlim([0, 256])
plt.title('Histogram of Color Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

# **TASK 1 :**

**Assignment : Apply Brightness**
* **1. Load an image.**
* **2. Increase the brightness of the image.**
* **3. Display the original and augmented images.**

```python
import cv2
import numpy as np
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

# Read the image using OpenCV
image = cv2.imread('/kaggle/input/labno8/image.jpg')

# Convert the image from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert the NumPy array to a PIL Image
pil_image = Image.fromarray(image_rgb)

# Enhance the brightness
enhancer = ImageEnhance.Brightness(pil_image)
bright_image = enhancer.enhance(1.5)  # Increase brightness by a factor of 1.5

# Convert the brightened PIL image back to a NumPy array
bright_image_np = np.array(bright_image)

# Plot the original and brightened images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(bright_image_np)
axes[1].set_title('Brightened Image')
axes[1].axis('off')

plt.show()
```

# **TASK 2 :**

**Assignment : Apply Another Edge Detection Technique**
* **1. Load an image.**
* **2. Convert the image to grayscale.**
* **3. Apply the Sobel edge detection technique to the grayscale image.**
* **4. Display the original grayscale image and the Sobel edge-detected image.**
* **5. Hint: Use the cv2.Sobel function in OpenCV to apply the Sobel operator.**

```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Load an image
image = cv2.imread('/kaggle/input/labno8/image.jpg')

# Step 2: Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 3: Apply the Sobel edge detection technique to the grayscale image
sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_edges = cv2.magnitude(sobel_x, sobel_y)

# Step 4: Display the original grayscale image and the Sobel edge-detected image
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_edges, cmap='gray')
plt.title('Sobel Edge-detected Image')
plt.axis('off')

plt.show()
```
